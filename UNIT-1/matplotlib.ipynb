# Matplotlib

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../ml_datasets/Salaries.csv")
```

```python
df['rank'].value_counts()
```

### Barchat

```python
df['rank'].value_counts().plot.bar(color=['orange', 'royalblue', 'green'])
plt.title("Rank Frequency")
plt.xlabel('Designation')
plt.ylabel('Count')
plt.ylim(0,300)
plt.grid(which='both')
# plt.minorticks_on()
# plt.xscale('log')
plt.show()
```

### Pie Chart

```python
# df['rank'].value_counts().plot.pie(autopct='%.2f')
df['salary'].value_counts().plot.pie(autopct='%.2f')
plt.show()
```

### HISTOGRAM

#### Based on mean of the column

```python
df['salary'].plot.hist()
plt.axvline(df['salary'].mean(),color='r')  # ax-vertical line on graph
plt.xlabel("salary")
plt.show()
```

```python
df['salary'].plot.hist(bins=3)
plt.axvline(df['salary'].mean(),color='r')  # ax-vertical line on graph
plt.xlabel("salary")
plt.grid()
plt.show()

# Longer tail on right side: Right Skewed
# No. of elements more than MEAN are more in number 
```

#### Based on Median (Box-plot)

```python
df['salary'].plot.box()
plt.xlabel("salary")
plt.grid()
plt.show()
```

## Scatter Plot

### Using 2 Variables

```python
plt.scatter(df['salary'], df['yrs.service'], c='darkblue')
plt.show()
```
