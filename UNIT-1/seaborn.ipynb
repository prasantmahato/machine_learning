## SEABORN package

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../ml_datasets/Salaries.csv")
# tips = sns.load_dataset("tips")
```

```python
sns.histplot(data=df, x='salary', kde=True, color='r')
```

```python
sns.boxplot(data=df, x="sex", y="salary")
plt.show()
```

```python
sns.boxplot(data=df, x="sex", y="salary",hue="rank")
plt.show()

## CONCLUSION: Female co-workers are paid less
```

```python
sns.countplot(data=df, x='rank')
plt.show()
```

```python
sns.scatterplot(data=df, x="yrs.service", y="salary", hue="sex")
plt.show()
```

```python
sns.lmplot(data=df, x="yrs.service", y="salary", hue="sex")
plt.show()
```

```python
sns.relplot(data=df, x="yrs.service", y="salary", hue="rank", style="sex", kind="line")
plt.show()
```

#### Regression Plot

```python
sns.regplot(data=df, x="yrs.service", y="salary")
plt.show()
```

```python
sns.regplot(data=df, x="yrs.service", y="yrs.since.phd")
plt.show()
```

```python
sns.barplot(data=df, x="rank", y="yrs.since.phd", hue="sex")    # x = categorical, y = numerical & hue = catergorical
plt.show()
```

```python
sns.catplot(data=df, x="rank", y="yrs.since.phd", hue="sex", kind="box")    # x = categorical, y = numerical & hue = catergorical
plt.show()
```

```python
sns.catplot(data=df, x="rank", y="salary", hue="sex", kind="swarm")    # x = categorical, y = numerical & hue = catergorical
plt.show()
```

```python
sns.catplot(data=df, x="rank", y="salary", hue="sex", kind="violin")
plt.show()
```

```python
sns.swarmplot(data=df, x="rank", y="salary", hue="sex")
plt.show()
```

```python
sns.lineplot(data=df, x=df.index, y="yrs.since.phd",hue="sex")
plt.show()
```

```python
sns.lineplot(data=df, y="yrs.since.phd", x="yrs.service", hue="sex")
plt.show()
```
