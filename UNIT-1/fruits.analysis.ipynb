### Research Question: Can apples and oranges be distinguished based on weight

```python
import pandas as pd
import seaborn as scs
import matplotlib.pyplot as plt
```

```python
scs.barplot(data=df, x="Weight", y="Size", hue="Class")
```

```python
scs.histplot(data=df, x="Weight", hue="Class", kde=True)
```

## Apple is heavier than oranges

###   Weight of oranges = 65 - 70 && Weight of apple = 70 - 75

```python
scs.catplot(data=df, x="Class", y="Weight", kind="box")
plt.grid()
```

#### Apples and oranges can be distuingished based on weight, Apple weighs heavier(70-75) and oranges weighs less in range(65-70)

```python
scs.catplot(data=df, x="Weight", y="Size", hue="Class", kind='box')
```
